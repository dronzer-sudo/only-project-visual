```python
# Q->C
# Logistic Regression
# Ho: Two variables are independent, Ha: Two variables are dependent
# based on p-value
# Hitters dataset
value = pd.read_excel('rushi proj.xlsx')
value.head()
value.set_index('Attrition', inplace = True)
value.dropna(inplace = True)

X = value['MonthlyIncome']
value['age is high'] = pd.cut(value['Age'], [value['Age'].min(), 40,
                                            value['Age'].max()],
                                            labels = ['L','H'])
```

```python
value['age is high'].head(10)
```

```python
y = value['age is high']=='H'
```

```python
y.head(10)
```

```python
import statsmodels.api as sm
X2 = sm.add_constant(X)
logit_model=sm.Logit(y,X2)
result=logit_model.fit()
print(result.summary())
```

```python
# Q->C
#here we monthly income split into two parameter below 7000 is  low and  above is high and then we check relation 
#between age and monthly income
value = pd.read_excel('rushi proj.xlsx')
value.head()
value.set_index('Attrition', inplace = True)
value.dropna(inplace = True)

X = value['Age']
value['MonthlyIncome'] = pd.cut(value['MonthlyIncome'], [value['MonthlyIncome'].min(), 7000,
                                            value['MonthlyIncome'].max()],
                                            labels = ['L','H'])
y = value['MonthlyIncome']=='H'
import statsmodels.api as sm
X2 = sm.add_constant(X)
logit_model=sm.Logit(y,X2)
result=logit_model.fit()
print(result.summary())
```

```python
# More thab Two independent samples (C->Q)
# One way F test - Anova
import pandas as pd
import numpy as np
from statsmodels.stats.proportion import proportions_ztest
from scipy.stats import ttest_ind
from scipy import stats



value = pd.read_excel('rushi proj.xlsx')   #here change rushi proj with
value['Attrition'] = value['Attrition'].astype('category')
value = value[['MonthlyIncome','Attrition']]
grps = pd.unique(value['Attrition'])
d_data = {grp:value['MonthlyIncome'][value.Attrition == grp] for grp in grps}
print(d_data)



```

```python
# More thab Two independent samples (C->Q)
# One way F test - Anova
import pandas as pd
cars = pd.read_csv('mtcars.csv')
cars['cyl'] = cars['cyl'].astype('category')
cars = cars[['mpg','cyl']]
grps = pd.unique(cars['cyl'])
d_data = {grp:cars['mpg'][cars.cyl == grp] for grp in grps}
print(d_data)
stats.f_oneway(d_data[4],d_data[6],d_data[8])
```

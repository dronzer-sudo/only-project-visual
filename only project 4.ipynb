```python
# Three predictors mpg ~ cyl
X_R1 = pd.get_dummies(rega['BusinessTravel'])
y_R1 = rega['MonthlyRate']
```

```python
linreg = LinearRegression().fit(X_R1, y_R1)
print("linear model intercept (b0): {}".format(linreg.intercept_))
print("linear model coeff (b1): {}".format(linreg.coef_))

y_hat = linreg.intercept_ + linreg.coef_[0]*X_R1.iloc[:,0]+linreg.coef_[1]*X_R1.iloc[:,1]+ linreg.coef_[2]*X_R1.iloc[:,2]
RSS = np.sum((y_R1 - y_hat)**2)
R2 = linreg.score(X_R1, y_R1)
n = X_R1.shape[0]
p = X_R1.shape[1]
adjR2 = 1 - (1-R2)*(n-1)/(n-p-1)
y_pred = linreg.predict(X_R1)
MSE = mean_squared_error(y_R1, y_pred)
RSE = np.sqrt(RSS/(n-p-1))

print('TSS: ', TSS)
print('RSS: ', RSS)
print('R-squared: ', R2)
print('Adjusted R-squared: ', adjR2)
print('RSE: ', RSE)

RSS_list.append(RSS)
R2_list.append(R2)
adjR2_list.append(adjR2)
RSE_list.append(RSE)
model_list.append('[BusinessTravel]')
```
